## 3. Generación de Insights

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Ejercicio 6: Análisis de Ventas

data = { 'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio'],
'Ventas': [15000, 18000, 17000, 14000, 20000, 19000],
'Costo Promocional': [2500, 3000, 2800, 2400, 3500, 3200] }

df = pd.DataFrame(data)

df
```

```python
# 1.	Genera un gráfico de líneas que muestre las ventas por mes.

plt.figure(figsize=(10, 6))
plt.plot(df['Mes'], df['Ventas'])
plt.title('Ventas por Mes')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.show();
```

```python
# 2.	Crea un gráfico de dispersión que muestre la relación entre ventas y costos promocionales.

plt.figure(figsize=(10, 6))
plt.scatter(df['Ventas'], df['Costo Promocional'])
plt.title('Relación entre Ventas y Costos Promocionales')
plt.xlabel('Ventas')
plt.ylabel('Costos Promocionales')
plt.show();
```

**3.	A partir de los gráficos, proporciona un insight sobre cómo los costos promocionales afectan las ventas.**

Se puede observar en el grafico de dispersion que efectivamente existe una correlacion positiva entre el costo promocional y las ventas, lo que significa que a mayor costo promocional mayor cantidad de ventas. Esto se evidencia en el grafico de lineas donde se ve un incremento de ventas del mes de Abril a Mayo y donde hubo un mayor costo promocional.

```python
# Ejercicio 7: Análisis de Productos Vendidos

data = { 'Día': ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes'],
 'Producto A': [50, 45, 60, 40, 55],
'Producto B': [30, 35, 25, 20, 40],
'Producto C': [70, 65, 75, 80, 85]}

df2 = pd.DataFrame(data)

df2
```

```python
# 1.	Crea un gráfico de barras apiladas que muestre la cantidad de productos vendidos por día.

plt.figure(figsize=(12, 8))
df2.set_index('Día').plot(kind='bar', stacked=True)
plt.title('Cantidad de Productos Vendidos por Día')
plt.xlabel('Día')
plt.ylabel('Cantidad Vendida')
plt.show();
```

```python
# 2.	Calcula la media y la desviación estándar de las ventas de cada producto.

df2.describe()
```

```python
media_a = df2['Producto A'].mean()

desviacion_estandar_a = df2['Producto A'].std()

print(f"La media de las ventas de Producto A es: {media_a}")
print(f"La desviación estándar de las ventas de Producto A es: {desviacion_estandar_a}")
```

```python
media_b = df2['Producto B'].mean()

desviacion_estandar_b = df2['Producto B'].std()

print(f"La media de las ventas de Producto A es: {media_b}")
print(f"La desviación estándar de las ventas de Producto A es: {desviacion_estandar_b}")
```

```python
media_c = df2['Producto C'].mean()

desviacion_estandar_c = df2['Producto C'].std()

print(f"La media de las ventas de Producto A es: {media_c}")
print(f"La desviación estándar de las ventas de Producto A es: {desviacion_estandar_c}")
```

**3.	Proporciona un insight sobre qué productos son más consistentes en términos de ventas y cuál tiene más variabilidad.**

De acuerdo con el grafico de barras el producto mas consistente en ventas seria el B ya que sus proporciones son parecidas cada dia mientras que el producto C prresenta mayor variabilidad.

```python
# Ejercicio 8: Relación entre Ventas y Publicidad

data = { 'Semana': ['Semana 1', 'Semana 2', 'Semana 3', 'Semana 4', 'Semana 5'],
'Ventas': [10000, 15000, 14000, 13000, 17000],
'Presupuesto Publicitario': [2000, 2500, 2200, 2100, 2800] }

df3 = pd.DataFrame(data)

df3
```

```python
# 1.	Crea un gráfico de dispersión que muestre la relación entre ventas y presupuesto publicitario.

plt.figure(figsize=(10, 6))
plt.scatter(df3['Ventas'], df3['Presupuesto Publicitario'])
plt.title('Relación entre Ventas y Presupuesto Publicitario')
plt.xlabel('Ventas')
plt.ylabel('Presupuesto Publicitario')
plt.show();
```

```python
# 2.	Calcula la correlación entre ambas variables.

df3[['Ventas', 'Presupuesto Publicitario']].corr()
```

**3.	Proporciona un insight sobre si el aumento en el presupuesto publicitario está relacionado con el aumento en las ventas.**

Se puede observar en el grafico de dispersion que efectivamente existe una correlacion positiva entre el presupuesto publicitario y las ventas, lo que significa que a mayor presupuesto publicitario mayor cantidad de ventas. Esto se evidencia una fuerte correlacion al calcular la correlacion donde el resultado es muy cercano a 1.
