**Gráfico de Línea**

```python
import matplotlib.pyplot as plt
import numpy as np


dias = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes']
ventas = [1000, 1500, 1200, 1800, 1600]

plt.plot(dias, ventas)
plt.title('Ventas semanales de un producto')
plt.xlabel('Días')
plt.ylabel('Ventas')

plt.show()
```

```python
plt.plot(dias, ventas,linestyle='--',color ='purple')
plt.title('Ventas semanales de un producto')
plt.xlabel('Días')
plt.ylabel('Ventas')
plt.grid(True)
plt.show()
```

**Gráfico de Barras**

```python
dia = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes']
ganancias = [600, 800, 700, 900, 800]

fig = plt.figure(figsize=(12, 6))

plt.bar(dia, ganancias)
plt.title('Ganancias semanales de un producto')
plt.xlabel('Días')
plt.ylabel('Ganancias')
plt.grid()
plt.show()
```

```python
fig = plt.figure(figsize=(12, 6))

plt.bar(dia, ganancias, color = 'green', width = 0.7)
plt.title('Ganancias semanales de un producto')
plt.xlabel('Días')
plt.ylabel('Ganancias')
plt.grid(axis='y')
plt.show()
```

**Histogramas**

```python
venta = [1000, 1500, 1200, 1800, 1600, 1300, 1700, 1400, 2000, 1100, 1900, 1800,
          2100, 1600, 1400, 1300, 1600, 1700, 1900, 1500, 1700, 1600, 1200, 1500,
          1300, 1400, 1800, 2100, 1100, 2000]

plt.hist(venta, bins=10, color='lime')
plt.title('Distribución de ventas')
plt.xlabel('Ventas')
plt.ylabel('Cantidad')
plt.grid(axis='y')
plt.show()
```

```python
plt.hist(venta, bins=8, color='indigo', alpha=0.7)
plt.title('Distribución de ventas')
plt.xlabel('Ventas')
plt.ylabel('Cantidad')
plt.grid(axis='y')
plt.show()
```

**Gráfico de Cajas**

```python
ganancia = [600, 800, 700, 900, 800, 750, 850, 720, 680]

plt.boxplot(ganancia)
plt.title('Distribución de ganancias')
plt.xlabel('Ganancia')
plt.ylabel('Cantidad')
plt.grid(axis='y')
plt.show()
```

```python
plt.boxplot(ganancia,patch_artist=True, boxprops=dict(facecolor='orange'))
plt.title('Distribución de ganancias')
plt.xlabel('Ganancia')
plt.ylabel('Cantidad')
plt.grid(axis='y')
plt.show()
```

**Gráfico de Torta**

```python
ventas_2 = [1000, 1500, 1200, 1800, 1600]
dias_2 = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes']

plt.pie(ventas_2, labels=dias_2)
plt.title('Ventas semanales de un producto')
plt.show()
```

```python
plt.pie(ventas_2, labels=dias_2, autopct='%1.1f%%', explode=(0, 0, 0, 0.1, 0))
plt.title('Ventas semanales de un producto')
plt.show()
```

**Gráfico de Dispersión**

```python
ventas_3 = [1000, 1500, 1200, 1800, 1600]
costos_3 = [400, 700, 500, 900, 800]

plt.scatter(ventas_3, costos_3, color = 'r' )
plt.title('Relación entre ventas y costos')
plt.xlabel('Ventas')
plt.ylabel('Costos')
plt.grid()
plt.show()
```

```python
ventas_4 = [1000, 1500, 1200, 1800, 1600]
ganancias_4 = [600, 800, 700, 900, 800]

plt.scatter(ventas_4, ganancias_4, color = 'green', s=100 )
plt.title('Relación entre ventas y costos')
plt.xlabel('Ventas')
plt.ylabel('Ganancias')
plt.grid()
plt.show()
```
