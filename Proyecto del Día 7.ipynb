### Objetivo
Crear un programa en Python que analice un conjunto de **datos de ventas de una tienda**. El programa debe realizar varias operaciones de Data Science para proporcionar información valiosa sobre las ventas de la tienda.

### Consigna

**Lectura de Datos**: 

```python
import pandas as pd
```

```python
ruta1 = r'C:\Users\steph\Documents\CURSOS\ANALISIS Y CIENCIA DE DATOS\PYTHON PARA DATA SCIENCE-UDEMY\Día 7\Datos_Ventas_Tienda.csv'
ventas1 = pd.read_csv(ruta1)
ventas1
```

```python
ruta2 = r'C:\Users\steph\Documents\CURSOS\ANALISIS Y CIENCIA DE DATOS\PYTHON PARA DATA SCIENCE-UDEMY\Día 7\Datos_Ventas_Tienda2.csv'
ventas2 = pd.read_csv(ruta2)
ventas2
```

**Fusión de Datos:**

```python
df_final = pd.concat([ventas1, ventas2], ignore_index = True)
df_final.tail()
```

**Tratamiento de Datos:**

```python
df_final['Fecha'] = pd.to_datetime(df_final['Fecha'])
df_final.head()
```

```python
df_final.info()
```

```python
df_final.describe()
```

**Análisis de Ventas:**

```python
# Producto mas vendido
producto_mas_vendido = df_final.groupby('Producto')['Cantidad'].sum()
producto_mas_vendido = producto_mas_vendido.sort_values(ascending=False)
print("El producto mas vendido es:" )
print(producto_mas_vendido.head(1))
```

```python
# Mes con mas ventas
meses = []

for f in df_final['Fecha']:
    meses.append(f.month)
    
df_final['Meses'] = meses
df_final.head()
    
```

```python
mes_con_mas_ventas = df_final.groupby('Meses')['Total Venta'].sum().sort_values(ascending= False).head(1)
print("El mes con mas ventas es: ")
mes_con_mas_ventas
```

**Datos Agrupados:**

```python
ventas_por_categorias = df_final.groupby('Producto')['Total Venta'].sum()
print("El total de ventas por cada categoria es: ")
ventas_por_categorias
```

**Guardar Resultados**

```python
df_final.to_csv(r'C:\Users\steph\Documents\CURSOS\ANALISIS Y CIENCIA DE DATOS\PYTHON PARA DATA SCIENCE-UDEMY\Día 7\Proyecto del Día 7.csv')
```
