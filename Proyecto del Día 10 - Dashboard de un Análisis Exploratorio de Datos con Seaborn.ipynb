***1. Cargar el dataset***

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
diamantes = sns.load_dataset("diamonds")
```

***2. Análisis exploratorio***

***3. Crear visualizaciones***

Relaciones entre variables numéricas: Precio y el Peso en Quilates de los Diamantes

```python
sns.set(style="whitegrid")

g = sns.relplot(data=diamantes,
                x="carat",
                y="price",
                kind="scatter",
                height=6,
                aspect=1.5)

g.fig.suptitle('Relación entre el Precio y el Peso en Quilates de los Diamantes',
               va='baseline',
               ha='center')

g.set_axis_labels('Peso en Quilates', 'Precio');
```

Distribuciones de variables numéricas : Precio

```python
g = sns.displot(data=diamantes,
                x="price",
                kind="hist",
                kde=True,
                color="blue",
                height=4,
                aspect=2)
g.fig.suptitle('Distribución del Precio de los Diamantes',
               va='baseline',
               ha='center')
g.set_axis_labels('Precio ($)', 'Frecuencia');
```

Distribuciones de variables numéricas : Peso

```python
g = sns.displot(data=diamantes,
                x="carat",
                kind="hist",
                kde=True,
                color="blue",
                height=4,
                aspect=2)
g.fig.suptitle('Distribución del Peso de los Diamantes',
               va='baseline',
               ha='center')
g.set_axis_labels('Carat', 'Frecuencia');
```

Comparaciones de variables numéricas agrupadas por categorías

```python
g = sns.catplot(data=diamantes,
                kind="box",
                x="cut",
                y="price",
                height=5,
                aspect=2,
                order=["Fair", "Good", "Very Good", "Premium", "Ideal"])

g.fig.suptitle('Comparación del Precio de los Diamantes por Calidad del Corte',
               va='baseline',
               ha='center')

g.set_axis_labels("Calidad del Corte",
                  "Precio ($)");
```

***4. Explorar relaciones específicas entre pares de variables***

```python
g = sns.jointplot(data=diamantes,
                  x="depth",
                  y="price",
                  kind="kde",
                  fill=True,
                  space=0,
                  color="purple")

g.set_axis_labels("Profundidad del Diamante (%)",
                  "Precio ($)")

g.fig.suptitle('Relación entre la Profundidad y el Precio de los Diamantes',
               va='baseline',
               ha='center',
               fontsize=16);
```

***5. Visualizar las relaciones entre todas las variables numéricas del dataset de manera conjunta***

```python
g = sns.pairplot(diamantes,
                corner=True)

g.fig.suptitle('Relaciones entre Variables Numéricas en el Dataset Diamonds',
               va='baseline',
               ha='center',
               fontsize=16);
```

# Interpretacion:

- En el tercer grafico se observa que la mayoría de los diamantes tienen pesos bajos, con un pico importante entre 0.2 y 0.5 quilates, donde la frecuencia llega a superar los 5,000 y a medida que aumenta el peso de los diamantes, la frecuencia de estos disminuye significativamente.

- En el sexto grafico se muestras las relaciones entre las distintas variables numericas entre las cuales las que mas muestra correlacion serian precio vs peso tal como se indico anteriormente.
